```python
from dotenv import load_dotenv, find_dotenv
import os

env_path = find_dotenv(usecwd=True)  # searches upward for .env
print("Found .env at:", env_path)

load_dotenv(env_path, override=True)

print("WHOOP_CLIENT_ID loaded:", bool(os.getenv("WHOOP_CLIENT_ID")))
print("WHOOP_CLIENT_SECRET loaded:", bool(os.getenv("WHOOP_CLIENT_SECRET")))
print("OPENAI_API_KEY loaded:", bool(os.getenv("OPENAI_API_KEY")))
print("WHOOP_REDIRECT_URI:", os.getenv("WHOOP_REDIRECT_URI"))
```
