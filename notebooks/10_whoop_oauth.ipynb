```python
from dotenv import load_dotenv
import os

load_dotenv(override=True)

required = ["WHOOP_CLIENT_ID", "WHOOP_REDIRECT_URI"]
for k in required:
    print(f"{k} loaded:", bool(os.getenv(k)))

print("WHOOP_REDIRECT_URI =", os.getenv("WHOOP_REDIRECT_URI"))
```

```python
import urllib.parse
import secrets
import os

WHOOP_CLIENT_ID = os.getenv("WHOOP_CLIENT_ID")
REDIRECT_URI = os.getenv("WHOOP_REDIRECT_URI", "https://oauth.pstmn.io/v1/callback")

SCOPES = "offline read:recovery read:sleep read:workout read:cycles"
AUTH_URL = "https://api.prod.whoop.com/oauth/oauth2/auth"

STATE = secrets.token_urlsafe(16)  # >= 8 chars, strong entropy
print("STATE (should come back after redirect):", STATE)

params = {
    "response_type": "code",
    "client_id": WHOOP_CLIENT_ID,
    "redirect_uri": REDIRECT_URI,
    "scope": SCOPES,
    "state": STATE,
}

auth_link = AUTH_URL + "?" + urllib.parse.urlencode(params)
print("\nOpen this URL, log in, approve scopes:\n")
print(auth_link)
```
